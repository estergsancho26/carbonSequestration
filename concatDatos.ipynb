```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Suponiendo que dfLR es tu DataFrame y ya está cargado

# Diccionario para convertir nombre de mes en español a número
meses_a_numero = {
    'Enero': '01', 'Febrero': '02', 'Marzo': '03', 'Abril': '04',
    'Mayo': '05', 'Junio': '06', 'Julio': '07', 'Agosto': '08',
    'Septiembre': '09', 'Octubre': '10', 'Noviembre': '11', 'Diciembre': '12'
}

# Convertir columnas 'año' y 'mes' a formato de fecha para facilitar la graficación
dfLR['Fecha'] = dfLR['año'].astype(str) + '-' + dfLR['mes'].map(meses_a_numero)

# Convertir la columna 'Fecha' a datetime
dfLR['Fecha'] = pd.to_datetime(dfLR['Fecha'], format='%Y-%m')

# Ordenar los datos por fecha
dfLR.sort_values('Fecha', inplace=True)

# Graficar la temperatura media mensual a lo largo del tiempo
plt.figure(figsize=(12, 8))  # Tamaño más grande para mejor visualización
plt.plot(dfLR['Fecha'], dfLR['TempMedia'], marker='o', linestyle='-', color='#6C8EBF')
plt.title('Evolución de la Temperatura Media Mensual (2010-2020)')
plt.xlabel('Fecha')
plt.ylabel('Temperatura Media (°C)')

# Formatear el eje X para mostrar las fechas más claramente
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Ajustar los ticks del eje Y para reflejar el rango real de los datos
plt.yticks(range(int(min(dfLR['TempMedia'])) - 4, int(max(dfLR['TempMedia'])) + 4, 2))

plt.grid(True)  # Añadir una cuadrícula para mejor visualización
plt.tight_layout()  # Ajusta automáticamente los parámetros de la subtrama para dar espacio al contenido

# Mostrar la gráfica
plt.show()
```
